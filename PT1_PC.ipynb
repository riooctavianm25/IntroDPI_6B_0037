```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bgr = cv2.imread('Knn.jpg')

img_bgr = cv2.resize(img_bgr, (400, 450))

print("Tipe data:", type(img_bgr))
print("Shape    :", img_bgr.shape)
print("Dtype    :", img_bgr.dtype)

print()
print("Artinya")
print(f"  Tinggi  = {img_bgr.shape[0]} pikses")
print(f"  Lebar   = {img_bgr.shape[1]} pikses")
print(f"  Channel = {img_bgr.shape[2]} (BGR)")

img_bgr = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(30, 10))
plt.imshow(img_bgr)
plt.title('Gambar bunga\n(Sudah dikonversi BGR ➡️ RGB)')
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

area_kecil = img_gray[200:208, 180:188]

print("=== MATRIKS f(x,y) - Area 8x8 Piksel ===")
print("Setiap angka = nilai intensitas 1 piksel (0=hitam, 255=putih)")
print()
print(area_kecil)
print()
print("Contoh baca:", f"f(180, 200) = {img_gray[200, 180]}")
print("Contoh baca:", f"f(181, 201) = {img_gray[201, 181]}")

fig, axes = plt.subplots(1,2, figsize=(12, 5))

axes[0].imshow(img_gray, cmap='gray', vmin=0, vmax=255)
axes[0].set_title("Gambar Grayscale\n(Setiap piksell = 1 angka)")
axes[0].axis('off')
rect = plt.Rectangle((180, 200), 8, 8, linewidth=2, edgecolor='red', facecolor='none')
axes[0].add_patch(rect)
axes[0].text(190, 195, 'area ini\ndizoom', color='red', fontsize=8)

axes[1].imshow(area_kecil, cmap='gray', vmin=0, vmax=255)
for i in range (8):
  for j in range (8):
    nilai = area_kecil[i,j]
    warna_teks = 'white' if nilai < 128 else 'black'
    axes[1].text(j, i, str(nilai), ha='center', va='center',
                 fontsize=9, color=warna_teks, fontweight='bold')
axes[1].set_title("Zoom 8x8 Piksel\nangka = nilai f(x,y)")
axes[1].axis('off')

plt.suptitle('SLIDE 4: Representasi Matematis f(x,y)', fontweight='bold')
plt.tight_layout()
plt.show
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

area_kecil_rgb = img_rgb[200:208, 180:188]

print("=== MATRIKS f(x,y) RGB - Area 8x8 Piksel ===")
print(area_kecil_rgb)
print()
print("Contoh baca:", f"f(180, 200) = [R, G, B] = {img_rgb[200, 180]}")
print("Contoh baca:", f"f(181, 201) = [R, G, B] = {img_rgb[201, 181]}")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].imshow(img_rgb)
axes[0].set_title("Gambar Berwarna")
axes[0].axis('off')

rect = plt.Rectangle((180, 200), 12, 12, linewidth=2, edgecolor='red', facecolor='none')
axes[0].add_patch(rect)
axes[0].text(190, 195, 'area ini\ndizoom', color='red', fontsize=8)

axes[1].imshow(area_kecil_rgb)

for i in range(8):
    for j in range(8):
        r, g, b = area_kecil_rgb[i, j]

        kecerahan = 0.299*r + 0.587*g + 0.114*b
        warna_teks = 'black' if kecerahan > 128 else 'white'

        teks_rgb = f"{r}\n{g}\n{b}"

        axes[1].text(j, i, teks_rgb,
                     ha='center', va='center',
                     fontsize=10, color=warna_teks, fontweight='bold')

axes[1].set_title("Zoom 8x8 (Nilai Matriks [R, G, B])")
axes[1].axis('off')

plt.suptitle('SLIDE 4: Representasi Matematis f(x,y) Citra RGB', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
h_g, w_g = img_gray.shape
channel_g = 1
bit_per_channel = 8

bit_gray = h_g * w_g * channel_g * bit_per_channel
byte_gray = bit_gray / 8
kb_gray = byte_gray / 1024
mb_gray = kb_gray / 1024

print("=== RINCIAN UKURAN GAMBAR GRAYSCALE ===")
print(f"Resolusi      : {w_g} x {h_g} piksel")
print(f"Channel       : {channel_g} (Hitam Putih)")
print(f"Ukuran (Bit)  : {bit_gray:,.0f} bits")
print(f"Ukuran (Byte) : {byte_gray:,.0f} Bytes")
print(f"Ukuran (KB)   : {kb_gray:,.2f} KB")
print(f"Ukuran (MB)   : {mb_gray:,.2f} MB")
```

```python
h_r, w_r, c_r = img_rgb.shape
bit_per_channel = 8

bit_rgb = h_r * w_r * c_r * bit_per_channel
byte_rgb = bit_rgb / 8
kb_rgb = byte_rgb / 1024
mb_rgb = kb_rgb / 1024

print("=== RINCIAN UKURAN GAMBAR RGB ===")
print(f"Resolusi      : {w_r} x {h_r} piksel")
print(f"Channel       : {c_r} (Red, Green, Blue)")
print(f"Ukuran (Bit)  : {bit_rgb:,.0f} bits")
print(f"Ukuran (Byte) : {byte_rgb:,.0f} Bytes")
print(f"Ukuran (KB)   : {kb_rgb:,.2f} KB")
print(f"Ukuran (MB)   : {mb_rgb:,.2f} MB")
```
